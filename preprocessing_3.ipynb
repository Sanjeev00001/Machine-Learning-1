```python
import pandas as pd
```

```python
dataset = pd.read_csv('/content/adult.csv', header = None, na_values = ' ?')
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
dataset.columns=['age', 'workclass', 'wgt' , 'edu', 'edu_num', 'marital_status', 'occup', 'relation', 'race' ,'sex' ,'gain', 'loss', 'hpw', 'country', 'income']
```

```python
dataset.head()
```

```python
## to find the null values
dataset.isnull().sum()
```

```python
# impute the data
# sklearn =  dataset is converted into dataframe
from sklearn.impute import SimpleImputer
si=SimpleImputer(strategy='most_frequent')
dataset1=si.fit_transform(dataset)
print(pd.DataFrame(dataset1).isnull().sum())
```

```python
dataset1=pd.DataFrame(dataset1)
print(dataset1.head())
```

```python
dataset1=pd.DataFrame(dataset1)
dataset1.columns=['age', 'workclass', 'wgt' , 'edu', 'edu_num', 'marital_status', 'occup', 'relation', 'race' ,'sex' ,'gain', 'loss', 'hpw', 'country', 'income']
print(dataset1.head())
```

```python
cols=['workclass', 'edu', 'marital_status','occup', 'relation', 'race', 'sex', 'country', 'income']
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in cols:
  dataset1[i]=le.fit_transform(dataset1[i])
print(dataset1.head())
```

```python
dataset1.info()
```

```python
cols1=['age','wgt', 'edu_num', 'gain', 'loss', 'hpw']
for i in cols1:
  dataset1[i] = dataset1[i].astype('int64')
dataset1.info()
```

```python
dataset1.describe()
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
dataset_std = sc.fit_transform(dataset1)
dataset_std=pd.DataFrame(dataset_std)
print(dataset_std.describe())
```
